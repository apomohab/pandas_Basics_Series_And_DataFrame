```python
#كيف نغير اسماء القيم المفقودة كما نريد - يمكننا ذلك عن طريق الاستدعاء للبيانات في البداية كالتالي 

data= pd.read_csv('data.csv' , na_values=[ '-' , 'na' , 'NAN']) #ستظهر البيانات المفقودة بهذه الأسماء فقط 

```

```python
#استبدال القيم المفقودة داخل آي عمود بقيمة أخري مثلا وضع -١ 

xchange = data['Salary'].fillna('-1' , inplace=True)

data.head()

#لاحظ الان في عمود الرواتب اصبحت القيمة المفقودة -١ 
#الدالة fillna هي المسئولة عن ذلك 

```

```python
#طريقة تغيير البيانات المفقودة داخل جميع الصفوف مرة واحدة

data.fillna('-1' , inplace=True)
data.head()
```

```python
#استبدال القيم المفقودة بالوسيط الحسابي لأي عمود بيانات

#اولا - الحصول علي الوسيط الحسابي 

get_mean = data['Age'].median()
get_mean
```

```python
#الان الوسيط الحسابي لعمود العمر = ٢٦ ، سنقوم بوقع هذا الرقم مكان القيم المفقودة في هذا العمود

add_mead = data['Age'].fillna(get_mean , inplace=True)
data[['Age']]

#القيمة المفقودة بالاسفل ظهرت -1 بناء علي العمليات السابقة ، ولكن يمكنك اختيار اي طربقة لتغيير القيم المفقودة كما تفضل
```

```python
#استبال القيم المفقودة بالقيمة الأكثر تكرارا داخل أي عمود - نحتاج أولا للتعرف علي القيم وعدد تكرارها كالتالي 

data['Team'].value_counts()

#سنجد ان الاسم New Orleans Pelicans قد ذكر ١٩ مرة 
#لذلك سنعوض القيم المفقودة بهذا الاسم داخل عمود Team 
#وهذه البيانات حصلنا عليها عن طريق الدالة value_counts
```
